```python
!pip install kagglehub[pandas-datasets]
import kagglehub
from kagglehub import KaggleDatasetAdapter
import seaborn as sns
import matplotlib.pyplot as plt

file_path = "personality_dataset.csv"
df = kagglehub.load_dataset(
    KaggleDatasetAdapter.PANDAS,
    "rakeshkapilavai/extrovert-vs-introvert-behavior-data",
    file_path,
)

print(df.head())
```

```python
print("\nGráfico 1: Histograma")

sns.histplot(data=df, x='Post_frequency')

plt.title('Histograma da Frequência de Postagens')

plt.show()
```

```python
print("\nGráfico 2: Curva de Densidade KDE")

sns.kdeplot(data=df, x='Post_frequency')

plt.title('Curva de Densidade da Frequência de Postagens')

plt.show()
```

```python
print("\nGráfico 3: Boxplot com Outliers")

sns.boxplot(data=df, x='Post_frequency')

plt.title('Boxplot da Frequência de Postagens (com outliers)')

plt.show()
```

```python
#Professor, era para ser sem outliers mas os dados não tinham uma distancia grande entre um e outro
print("\nGráfico 4: Boxplot (*sem outliers")

sns.boxplot(data=df, x='Post_frequency', showfliers=False)

plt.title('Boxplot da Frequência de Postagens (sem outliers)')

plt.show()
```
